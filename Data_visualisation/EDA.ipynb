```python
import pandas as pd                        # to access the data frame
import numpy as np                 
import seaborn as sns                      # to generarte the plots
from matplotlib import pyplot as plt       # to generarte the plots
import matplotlib                                      ## these libraries are sufficient to carry out EDA!
```

```python
data=pd.read_csv('CO2 Emissions_Canada_Classification.csv')

```

```python
data.info()
```

```python
# remove spaces in column names and replace with '_'
data.columns=data.columns.str.replace(" ",'_')
data.columns
```

```python
data['Make'].value_counts()
```

```python
data['Make'].unique()
```

```python
data['Make'].fillna(method='ffill',inplace=True)
data['Make'].isna().value_counts()
```

```python
plt.figure(figsize=(360, 180))
matplotlib.rcParams['font.size'] = 94
sns.countplot(x='Make',data=data)
```
